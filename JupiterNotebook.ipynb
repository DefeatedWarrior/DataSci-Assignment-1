```python
#Setup All Dependancies

import pandas as pd

import matplotlib.pyplot as plt

```

```python
#Open all data as students, myphone & userbehavior

students = pd.read_csv("data/Students_Social_Media_Addiction.csv")

myphone = pd.read_csv("data/MyPhone.csv")

userbehavior = pd.read_csv("data/user_behavior_dataset.csv")

students.head()
myphone.head()
userbehavior.head()
```

```python
plt.figure()

plt.scatter(
    myphone["Notifications"],
    myphone["Times Opened"],
    s=myphone["Day Duration (Minutes)"],  # bubble size

)

plt.xlabel("Notifications")
plt.ylabel("Times Opened")
plt.title("Notifications vs Times Opened (Bubble = Time Spent)")

plt.show()
```

```python
# Group by App
app_usage = myphone.groupby("AppClean")["Day Duration (Minutes)"].sum()

plt.figure()

plt.bar(app_usage.index, app_usage.values, color="red")

plt.xticks(rotation=90)
plt.xlabel("App")
plt.ylabel("Total Minutes")
plt.title("Total Screen Time by App (Over 10 Days)")

plt.show()
```

```python
# My total usage (convert minutes to hours & Divide by 10 days average)
my_total_hours = myphone["Day Duration (Minutes)"].sum() / 60 / 10

# Student average
avg_student_usage = students["Avg_Daily_Usage_Hours"].mean()

labels = ["(Wil)Liam", "Average Student"]
values = [my_total_hours, avg_student_usage]

plt.figure()

plt.bar(labels, values, color="green")

plt.ylabel("Hours per Day")
plt.title("(Wil)Liam vs Average Student Screen Time")

plt.show()
```

```python
plt.figure()

plt.scatter(
    students["Avg_Daily_Usage_Hours"],
    students["Mental_Health_Score"]
)

plt.xlabel("Avg Daily Usage (Hours)")
plt.ylabel("Mental Health Score")
plt.title("Social Media Usage vs Mental Health")

plt.show()
```
